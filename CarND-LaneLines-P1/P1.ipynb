Below are some helper functions to help get you started. They should look familiar from the lesson!

```python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def get_slopes_and_lines(lines):
    """
    Get the left and right lines and slopes
    """
    left_slopes = []
    left_lines = []
    right_slopes = []
    right_lines = []
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = ((y2-y1)/(x2-x1))
            #print(slope)
            if slope > 0: #right side of lane
                left_slopes.append(slope)
                left_lines.append(line)
            else:
                right_slopes.append(slope)
                right_lines.append(line)
    return left_slopes, left_lines, right_slopes, right_lines


def get_clean_lines(left_slopes, left_lines, right_slopes, right_lines):
    """
    Remove lines with slopes outside one standard deviation of mean
    """
    right_mean = np.mean(right_slopes)
    right_std = np.std(right_slopes)
    right_lines_clean = [right_lines[i] for i in range(len(right_lines)) if 
                        right_slopes[i] <= right_mean + right_std and right_slopes[i] >= right_mean - right_std]
    
    left_mean = np.mean(left_slopes)
    left_std = np.std(left_slopes)
    left_lines_clean = [left_lines[i] for i in range(len(left_lines)) if 
                        left_slopes[i] <= left_mean + left_std and left_slopes[i] >= left_mean - left_std]
    return right_lines_clean + left_lines_clean


def get_avg_points(clean_lines):
    """
    Average points from all lines provided, splitting between left and right
    """
    avg_left_x1 = 0
    avg_left_y1 = 0
    avg_left_x2 = 0
    avg_left_y2 = 0
    n_left = 0
    
    avg_right_x1 = 0
    avg_right_y1 = 0
    avg_right_x2 = 0
    avg_right_y2 = 0
    n_right = 0
    for line in clean_lines:
        for x1,y1,x2,y2 in line:
            slope = ((y2-y1)/(x2-x1))
            if slope > 0: #right side of lane
                avg_right_x1 = avg_right_x1 + x1
                avg_right_y1 = avg_right_y1 + y1
                avg_right_x2 = avg_right_x2 + x2
                avg_right_y2 = avg_right_y2 + y2
                n_right = n_right + 1
            else:
                avg_left_x1 = avg_left_x1 + x1
                avg_left_y1 = avg_left_y1 + y1
                avg_left_x2 = avg_left_x2 + x2
                avg_left_y2 = avg_left_y2 + y2
                n_left = n_left + 1
    
    avg_left_x1 = int(avg_left_x1 / n_left)
    avg_left_y1 = int(avg_left_y1 / n_left)
    avg_left_x2 = int(avg_left_x2 / n_left)
    avg_left_y2 = int(avg_left_y2 / n_left)
    
    avg_right_x1 = int(avg_right_x1 / n_right)
    avg_right_y1 = int(avg_right_y1 / n_right)
    avg_right_x2 = int(avg_right_x2 / n_right)
    avg_right_y2 = int(avg_right_y2 / n_right)
    return avg_left_x1, avg_left_y1, avg_left_x2, avg_left_y2, \
        avg_right_x1, avg_right_y1, avg_right_x2, avg_right_y2
    
    
def get_interpolated_points(avg_left_x1, avg_left_y1, avg_left_x2, avg_left_y2,
        avg_right_x1, avg_right_y1, avg_right_x2, avg_right_y2, imshape):
    """
    Interpolate lines to start at bottom of image and extend a fixed distance
    """
    left_slope = ((avg_left_y2-avg_left_y1)/(avg_left_x2-avg_left_x1))
    right_slope = ((avg_right_y2-avg_right_y1)/(avg_right_x2-avg_right_x1))
    left_b = avg_left_y1 - (left_slope * avg_left_x1)
    right_b = avg_right_y1 - (right_slope * avg_right_x1)
    
    right_y1 = left_y1 = imshape[1]
    left_x1 = int((left_y1 - left_b) / left_slope)
    right_x1 = int((right_y1 - right_b) / right_slope)
    
    # hard coded y2 value that seemed to work for this camera angle.
    # if had different camera angles would need to think of more robust method
    min_y2 = 335
    right_y2 = left_y2 = min_y2
    left_x2 = int((left_y2 - left_b) / left_slope)
    right_x2 = int((right_y2 - right_b) / right_slope)
    return right_x1, right_y1, right_x2, right_y2, left_x1, left_y1, left_x2, left_y2 

    
def draw_lines(img, lines, color=[255, 0, 0], thickness=8):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """

    left_slopes, left_lines, right_slopes, right_lines = get_slopes_and_lines(lines)
    clean_lines = get_clean_lines(left_slopes, left_lines, right_slopes, right_lines)
    avg_left_x1, avg_left_y1, avg_left_x2, avg_left_y2, \
        avg_right_x1, avg_right_y1, avg_right_x2, avg_right_y2 = get_avg_points(clean_lines)
    right_x1, right_y1, right_x2, right_y2, left_x1, left_y1, left_x2, left_y2 \
        = get_interpolated_points(avg_left_x1, avg_left_y1, avg_left_x2, avg_left_y2,
        avg_right_x1, avg_right_y1, avg_right_x2, avg_right_y2, img.shape)
    
    cv2.line(img, (right_x1, right_y1), (right_x2, right_y2), color, thickness)
    cv2.line(img, (left_x1, left_y1), (left_x2, left_y2), color, thickness)
    

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((*img.shape, 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
```

```python
def process_image(image):
    # NOTE: The output you return should be a color image (3 channel) for processing video below
    # TODO: put your pipeline here,
    # you should return the final output (image with lines are drawn on lanes)
    
    blur_filter_size = 5
    low_canny = 30
    high_canny = 150
    rho = 1
    theta = np.pi/180 
    threshold = 35
    min_line_length = 5
    max_line_gap = 40
    region_height_pct = 0.58

    imshape = image.shape
    gray = grayscale(image)
    blurred = gaussian_blur(gray, blur_filter_size)
    canny_img = canny(blurred, low_canny, high_canny)
    vertices = np.array([[(0,imshape[0]),(imshape[1]/2, 
                                          imshape[0]*region_height_pct), 
                          (imshape[1]/2, imshape[0]*region_height_pct), (imshape[1],imshape[0])]], dtype=np.int32)
    region = region_of_interest(canny_img, vertices)
    hough_img = hough_lines(region, rho, theta, threshold, min_line_length, max_line_gap)
    result = weighted_img(hough_img, image)

    return result
```

My current pipeline converts the image to gray scale, blurs the image, runs Canny edge detection, subsets the image to the region of interest (the triangle in front of the car), and then finds the lines using a Hough transform. From the lines returned I clean up the lines by removing any with slopes outside one standard deviation. I then average the points for each side and extrapolate starting from the bottom of the image up to a y value of 335.

One obvious way I could make my algorithm more robust is to remove the hard coded y value of 335. This works well given the current placement of the camera on the car, but might not always works. I could perhaps improve this by instead specifying a length that I want my line to be. My current algorithm is likely to fail when the lane lines are not in sharp contrast to the road. For example, in the challenge video part of the road becomes somewhat white.
